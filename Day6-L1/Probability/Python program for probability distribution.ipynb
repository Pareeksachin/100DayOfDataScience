##### Normal(Gaussian) distribution

Normal (Gaussian) Distribution:
		=> The normal distribution is characterized by a bell-shaped curve and is widely used due to its symmetry and well-understood properties.
		=> Real-world scenario: Human height, where the distribution is centered around the mean height with most people clustered near the average.

PDF(Probability Density Function ):
 
he Probability Density Function (PDF) represents the probability distribution of a continuous random variable. It describes the likelihood of the random variable taking on a particular value or falling within a certain range of values

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = np.random.normal(loc=0,scale=1,size = 1000)

#loc = mean of the normal distribution
#scale = standard deviation of the normal distribution. which means the spread or width of the distribution
#size = number of samples

#plot histogram 

plt.hist(data,bins=30,density = True,alpha = 0.6,color='b')

# bins = number of bins(intervals) into which the data is divided for the histogram
# density = normalize the histogram so that the area under the histogram equals 1
# alpha = control the transparency of the histogram bars

# plot probability density function(PDF)
#plt.show()
xmin,xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p = np.exp(-0.5*x**2)/np.sqrt(2*np.pi)
plt.plot(x,p,'k',linewidth=2)
plt.title("Normal Distribution")
plt.show()
```

#### Uniform Distribution

```python
low = 0 
high = 10
size = 1000 # number of samples

#Generate random number from a uniform distribution
data = np.random.uniform(low,high,size)

#plot histogram
plt.hist(data,bins=30,density=True,alpha = 0.6,color='b')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Uniform Distribution')

plt.show()
```

#### Exponential Distribution

```python
data = np.random.exponential(scale=1.0,size =1000)

#plot histogram
plt.hist(data,bins=30,density=True,alpha=0.6,color='b')

plt.show()
```

#### Gamma Distribution

```python
gamma_samples = np.random.gamma(shape=2.0,scale=2.0,size = 1000)

plt.hist(gamma_samples,bins = 30,density=True,alpha = 0.6,color='b')

plt.title('Gamma Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

#### beta distribution

```python
# Parameters for beta distribution
alpha = 2.0
beta = 5.0
size = 1000

# Generate random samples from beta distribution
beta_samples = np.random.beta(alpha, beta, size)

# Plot histogram of beta distribution
plt.hist(beta_samples, bins=30, density=True, alpha=0.6, color='r')
plt.title('Beta Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define different alpha and beta values
alpha_values = [0.5, 1, 2, 5]
beta_values = [0.5, 1, 2, 5]

# Generate x values for plotting
x = np.linspace(0, 1, 100)

# Plot beta distributions for different alpha and beta values
plt.figure(figsize=(10, 6))
for alpha in alpha_values:
    for beta in beta_values:
        # Calculate PDF for the current alpha and beta values
        y = beta.pdf(x, alpha, beta)
        # Plot PDF
        plt.plot(x, y, label=f'α={alpha}, β={beta}')

plt.title('Beta Distribution with Different α and β Values')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
