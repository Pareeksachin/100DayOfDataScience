#### Scatter plot

```python
#Scatter plots are used to visualize the relationship between two continuous variables. They help identify patterns, trends, and potential outliers in the data.
# Use Cases: Exploring correlations, showcasing distributions, highlighting clusters or groups in data

#generate some data

#This code will generate a different random number each time it is run, but if you run it multiple times 
#with the same seed, it will always generate the same random number. This is because the seed tells the 
#random number generator where to start in the sequence of random numbers.
np.random.seed(42)
x = np.random.rand(50)
y = 2*x + 1 + 0.1*np.random.randn(50) # linear relation with some noise

plt.scatter(x,y,color = 'blue',marker = 'o',label = 'Data Points')

# add lables and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")

#show the legend
plt.legend()

#display plot
plt.show()
```

#### Bar plot

```python
#Purpose: Bar plots are effective for visualizing and comparing the values of different categories or groups. They provide a clear representation of individual values or aggregated statistics.
#use cases : comparing sales figures across diff products , showing frequencies of categories , displaying results of experiments or surveys.

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [15,24,18,30]

#create a bar plot
plt.bar(categories,values,color = 'green',alpha=0.7)
# alpha parameter controls the transparency of the bars. A value of alpha=0.7 means that the bars will be 70% opaque


# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')

# Display the plot
plt.show()
```

#### Pie Chart

```python
#Show the composition or distribution of parts within a whole.
#Highlight the proportion of each category in relation to the total.

#Use Cases:

#Displaying the percentage breakdown of different categories in a dataset.
#Illustrating the distribution of budget allocation in different expense categories.
#Communicating the market share of different products or services.

sizes = [30,20,15,35]
labels = ['Category A', 'Category B', 'Category C', 'Category D']
explode = (0.1,0,0,0) # Explode the first slice (Category A)

#pie chart
plt.pie(sizes,labels = labels,autopct = '%1.1f%%',explode= explode,colors = ['red','green','blue','orange'],startangle=50)
#explode=explode: A tuple or array specifying the fraction of the radius with which to offset each wedge. In this case, (0.1, 0, 0, 0) means exploding the first wedge (Category A) by 10% of the radius.
#autopct = A string formatting specification for displaying the percentage contribution of each wedge on the pie chart
#startangle = The angle by which the start of the pie chart is rotated, measured in degrees
#add title 
plt.title("Pie Chart")

plt.show()
```

#### Histogram

```python
#Visualize the distribution of a continuous variable.
#Illustrate the frequency or count of data points within specific ranges (bins).
# usecases : 
#Analyzing the distribution of exam scores in a class.
#Understanding the distribution of heights or weights in a population.
#Exploring the frequency of sales amounts in different price ranges.

# Generate example data
np.random.seed(42)
data = np.random.normal(size=1000)

# Create a histogram
plt.hist(data, bins=20, color='purple', edgecolor='black', alpha=0.7)

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Display the plot
plt.show()
```

## Intermediate Level

#### Advanced Line Plots

```python
#Visualize and compare multiple datasets or trends on a single plot
#use cases : Analyzing the stock prices of multiple companies on the same chart.

# Example data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 2, 1, 2, 1]

# Plotting multiple lines with custom legends and labels
plt.plot(x, y1, label='Line 1', color='blue', linestyle='-', marker='o')
plt.plot(x, y2, label='Line 2', color='green', linestyle='--', marker='s')

# Adding labels and legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Advanced Line Plot')
plt.legend()

# Display the plot
plt.show()
```

#### Error Bars

```python
#Visualize uncertainties or variability in data points
#use cases : 
#displaying confidence intervals around mean values
#Representing measurement uncertainties in scientific experiments.
#Communicating variability in financial data.

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
errors = [0.5, 0.3, 0.7, 0.4, 0.6]  # Example error values

#plotting with error bars

plt.errorbar(x,y,yerr = errors,fmt='o-',color='purple',ecolor='red',capsize = 5)
#yerr=errors: The error values associated with each data point along the y-axis. This parameter adds vertical error bars to the plot.
#capsize=5: The size of the caps at the end of the error bars.
#ecolor='red': The color of the error bars.
#fmt='o-': The format of the plot, specifying the marker ('o' for circles), line style ('-' for solid line), and color ('purple' for the line color)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bars')

# Display the plot
plt.show()
```

#### Annotations and Text:


```python
#Purpose: Highlight specific points or features in the plot with descriptive text.
#use cases : 
#Pointing out outliers or interesting observations.
#Providing additional information about data points.
#Enhancing the storytelling aspect of the plot.

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

#plotting the data
plt.plot(x,y,marker='o',color='blue')

# adding text annotation

plt.text(3,6,'Important Point',fontsize = 10,color='red',ha='center')

# adding arrow notation : Direct attention to specific regions or trends in the plot

plt.annotate('Start',xy=(1,2),xytext = (1.5,4),arrowprops=dict(facecolor='green',shrink=0.05))
#'Start : displayed as annotation'
#xytext=(1.5, 4) specifies the position where the text ('Start') will be placed
#xy=(1, 2) sets the point where the annotation arrow will point
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Annotations and Text')

plt.show()
```

#### Image and colormaps

```python
#Purpose of Image plots: Visualizing and analyzing image data using Matplotlib.
#use cases : Displaying photographs, satellite images, or medical imaging data.
#Inspecting pixel intensities or patterns in an image

#purpose of colormaps : Enhancing visualizations by mapping data values to colors.
#use cases: Highlighting gradients or variations in datasets.
#Improving the interpretability of heatmaps or surface plots.

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Displaying an image
img = mpimg.imread('Lion.jpg')
plt.imshow(img)
plt.axis('off')  # Turn off axis labels
plt.title('Image Display Example')
plt.show()
```

#### 3D Plots:


```python
#Representing and visualizing data in three-dimensional space
#use cases : Plotting complex mathematical functions.

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Creating a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('3D Plot Example')
plt.show()
```

#### Heatmaps

```python
#Purpose: Visualizing matrix-like data with color intensity.
#Use cases : Displaying correlation matrices in statistics.
#Analyzing spatial patterns in scientific research.
# Creating a heatmap
data = np.random.random((10, 10))
plt.imshow(data, cmap='hot', interpolation='nearest')
#cmap='hot': The colormap used for mapping scalar data to colors. In this case, the 'hot' colormap is used, which ranges from black to red to yellow.
#interpolation='nearest': The method used for interpolation between pixels when displaying the image.

plt.colorbar(label='Intensity')
plt.title('Heatmap Example')
plt.show()
```

#### Polar Plots

```python
#Purpose: Representing data in a circular coordinate system.
#use cases : Analyzing cyclic patterns, such as wind direction.
#Displaying data with angular relationships.

# Creating a polar plot
theta = np.linspace(0, 2*np.pi, 100)
r = np.sin(3*theta)
plt.polar(theta, r, label='Polar Plot')
#theta: The array of angles in radians at which the data points are located around the polar plot.
#r: The radial distance from the origin for each corresponding angle theta. In this case, it's calculated as np.sin(3*theta), generating a sinusoidal pattern.
#
plt.title('Polar Plot Example')
plt.legend()
plt.show()
```
